```python
#plt.quiver([0, 0, 0], [0, 0, 0], [1, -2, 4], [1, 2, -7], angles='xy', scale_units='xy', scale=1)
#X,Y,U,V, X,Y donde inicia, y U,V donde termina la frecha

def v(x0,x1):
    return np.array([x0,x1])

def eucdis0(x0,x1):
    return np.linalg.norm(x0-x1)

def eucdis1(x0,x1):
    return np.sum( (x0-x1)**2 )**(1/2)
def cosAng(x0,x1):
    return (x0@x1) / (np.linalg.norm(x1) * np.linalg.norm(x1))
def Ang(x0,x1):
    return np.arccos(cosAng(x0,x1) )

#xi = v(1,3)
xi = v(2.9,2)
xj = v(3,2)
print ("Euclidean distance :")
print (eucdis0(xi,xj) )
print (eucdis1(xi,xj) )
print ("Dot product:")
print (xi@xj)#print ( np.sum(xi*xj) )
print ("Cosin of the angle:")
print ( cosAng(xi,xj) )
#print ("The angle:")
#print ( np.arccos(cosAng(xi,xj)) )

x = np.array([xi, xj]).T
ixi = np.zeros(len(x))
iyi = np.zeros(len(x))

colors = np.random.choice(["r","g","c","m"],len(x),replace=False)
plt.quiver(ixi, iyi, x[0], x[1], angles='xy', scale_units='xy', scale=1,color=colors)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```python
x1 = np.array([0,0,1,1])
x2 = np.array([0,1,0,1])
output = np.array([0,0,0,1])

w = np.random.normal(0,5,3)

net  = w[0]*x1 + w[1]*x2

def f_Activation_Signo(val):
    if val>0:
        return 1
    else:
        return 0
fsigno = np.vectorize(f_Activation_Signo)

fsigno(net)
```

```python
eta = .1
th = .5
for i in range(100):
    for x1i,x2i,y in zip(x1,x2,output):
        net = w[0]*x1i + w[1]*x2i + w[2]
        y_hat = fsigno(net)
        e = y - y_hat
        #WidrowHoff delta rule,
        w[0] = w[0] + eta*x1i*e
        w[1] = w[1] + eta*x2i*e
        w[2] = w[2] + eta*1*e
```

```python
from random import choice 
from numpy import array, dot, random 

unit_step = lambda x: 0 if x < 0 else 1 
training_data = [ (array([0,0,1]), 0), 
                  (array([0,1,1]), 1), 
                  (array([1,0,1]), 1), 
                  (array([1,1,1]), 1), ]

w = random.rand(3) 
errors = [] 
eta = 0.2 
n = 100 

for i in range(n): 
    x, expected = choice(training_data) 
    result = dot(w, x) 
    error = expected - unit_step(result) 
    errors.append(error) 
    w += eta * error * x 
    
    
for x, _ in training_data: 
    result = dot(x, w) 
    print("{}: {} -> {}".format(x[:2], result, unit_step(result)))


```

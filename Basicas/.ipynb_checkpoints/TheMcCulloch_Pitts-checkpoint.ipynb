# The McCulloch-Pitts Neuron Model

# One-Neuron Perceptron

The perceptron, also referred to as a McCulloch–Pitts neuron or linear threshold
gate, is the earliest and simplest neural network model. Rosenblatt used a single-layer perceptron **for the classification of linearly separable patterns**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda th0,th1,x: th1*x + th0

x = np.arange(0,10)
y_org = f(1,.5,x) + np.random.normal(0,1,len(x))
clase = x.copy()
clase[x <4] =0
clase[x >=4] =1

mark = ["o","x"]
color = ["r","b"]
for _x, _y,_c in zip(x, y_org,clase):
        plt.scatter(_x, _y, marker=mark[_c], c=color[_c])
```

```python
def logistic(x):
    beta=1
    return 1/(1 + np.exp(-beta*x ))
```

```python
w = np.random.normal(0,10,2)
#threshold = th
th = .5
```

```python
X = np.array([x,y_org]).T
```

```python
net = [ w@x- th for x in X]
y_hat = [ logistic(net) for x in X]
#net = [ np.tanh(w@x- th) for x in X]
```

```python
y_hat = [ logistic(w@x- th) for x in X]
e = y_org - y_hat
e
```

```python
eta=.001
w = np.random.normal(0,10,2)
th = 1
```

```python
for i in range(1000):
    y_hat = [ logistic(w@x- th) for x in X]
    #y_hat = [ np.sign(w@x- th) for x in X]
    e = clase - y_hat
    w = w + eta * x @ e

print (w)
```
